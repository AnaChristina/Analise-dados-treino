# Analisando dados da agricultura

## Importando bibliotecas üìö

```python
import pandas as pd 
import numpy as np
```

## Carregando Dados üé≤

```python
# df = data frame, df √© muito generico, mas como so estou usando 1 n√£o vejo problemas em utilizar esse nome para a variavel;
df = pd.read_csv('C:\\Users\\anach\\OneDrive\\√Årea de Trabalho\\analise de dados\\agricultural_raw_material.csv')
```

## Explorando os dados üïµÔ∏è‚Äç‚ôÄÔ∏è

```python
# info para conseguir entender as colunas e sobre oq se trata estes dados.
# isnull para ver se h√° registros nulos.
# head() mostra apenas os 5 primeiros registros. 
df.info
df.isnull().sum()

```

## Lidando com dados ausentes, incorretos e inv√°lidos

```python
# Replacing %, "," and "-"      substitui. 
df = df.replace('%','', regex=True)
df = df.replace(',','', regex=True)
df = df.replace('-','', regex=True)
df = df.replace('', np.nan)
df = df.replace('MAY90', np.nan)

#Dropping rows with NaN values ( Apagando linhas com valores Nulos)
df = df.dropna()

#Check to see if all NaN values are resolved
df.isnull().sum()

#Converting data type to float
lst = ['Coarse wool Price', 'Coarse wool price % Change',
       'Copra Price', 'Copra price % Change', 'Cotton Price',
       'Cotton price % Change', 'Fine wool Price', 'Fine wool price % Change',
       'Hard log Price', 'Hard log price % Change', 'Hard sawnwood Price',
       'Hard sawnwood price % Change', 'Hide Price', 'Hide price % change',
       'Plywood Price', 'Plywood price % Change', 'Rubber Price',
       'Rubber price % Change', 'Softlog Price', 'Softlog price % Change',
       'Soft sawnwood Price', 'Soft sawnwood price % Change',
       'Wood pulp Price', 'Wood pulp price % Change']
df[lst] = df[lst].astype("float")
df.dtypes
```

```python
df.head()
```

## Colunas de data e horacolunas
formatando a coluna datetime e definindo-a como √≠ndice para o conjunto de dados.

```python
df.Month = pd.to_datetime(df.Month.str.upper(), format="%b%y", yearfirst=False)
# Indexing month
df= df.set_index('Month')
```

```python
df.head()
```

## An√°lise Explorat√≥ria e Visualiza√ß√£o üìäüìà
importar bibliotecas matplotlib e seaborn para gerar os graficos

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# passando um padr√£o de como vai ser os gr√°ficos
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

**Mapa de Calor**

```python
#raw-materials list
raw_data=['Coarse wool Price','Copra Price', 'Cotton Price', 'Fine wool Price', 'Hard log Price', 'Hard sawnwood Price', 'Hide Price', 'Plywood Price', 'Rubber Price', 'Soft sawnwood Price', 'Wood pulp Price']

#getting the correlation matrix
corrmat = df[raw_data].corr()

#setting the size of plot
fig = plt.figure(figsize = (12, 9))

#masking the upper traingle part since matrix is symmetric(repetitive)
mask = np.triu(np.ones_like(corrmat, dtype=bool))
sns.heatmap(corrmat, vmax = .8, mask=mask, square = True, annot = True)

plt.show()
```

<p> O Mapa de Calor mostra a correla√ß√£o entre as mat√©rias-primas maior o valor correlacionado maior chance de serem duas mat√©rias-primas relacionadas, mas n√£o necessariamente. 
</p>  

Explore uma ou mais colunas tra√ßando um gr√°fico abaixo e adicione alguma explica√ß√£o sobre ele Matriz de Correla√ß√£o

```python
plt.figure(figsize=(30,15))
changelist=['Copra price % Change','Softlog price % Change','Rubber price % Change','Cotton price % Change','Coarse wool price % Change','Fine wool price % Change','Hard log price % Change','Hard sawnwood price % Change','Hide price % change','Plywood price % Change','Soft sawnwood price % Change','Wood pulp price % Change']

#Generate a correlation matrix for the whole dataset
corrMatrix = df[changelist].corr()
sns.heatmap(corrMatrix, annot = True)
plt.show()
```

 A partir deste r+gr√°fico podemos dizer que quase n√£o h√° rela√ß√£o entre a varia√ß√£o percentual dos pre√ßos das mat√©rias-primas
 - o valor negativo implica que duas vari√°veis est√£o negativamente correlacionadas (um aumento, outro decr√©scimo)
 - zero implica nenhuma rela√ß√£o
 - caso contr√°rio, maior o valor maior a chance de rela√ß√£o pre√ßos e seus gr√°ficos de % de mudan√ßa

### Resolvendo Quest√µes üéØ

#### Q1: Descubra a varia√ß√£o normal do pre√ßo de cada mat√©ria-prima

```python
# changelist=['Copra price % Change','Softlog price % Change','Rubber price % Change','Cotton price % Change','Coarse wool price % Change','Fine wool price % Change','Hard log price % Change','Hard sawnwood price % Change','Hide price % change','Plywood price % Change','Soft sawnwood price % Change','Wood pulp price % Change']
for i in range(len(changelist)):   ## para cada registro, ele faz um grafico diferente.
    plt.figure(figsize=(12,12))
    df[changelist[i]].hist(figsize=(11,9), linewidth=1)
    plt.xlabel('% Change') ## eixo X legenda
    plt.ylabel('count') ## eixo Y legenda
    plt.legend(changelist[i:], loc='upper center', bbox_to_anchor=(1.2,1))
    
```

Podemos observar que a maioria das mat√©rias-primas tem % de mudan√ßa frequente ideal inferior a 5%

#### Q2: Encontre a mat√©ria-prima que tem o menor pre√ßo ao longo dos anos

```python
# raw_data=['Coarse wool Price','Copra Price', 'Cotton Price', 'Fine wool Price', 'Hard log Price', 'Hard sawnwood Price', 'Hide Price', 'Plywood Price', 'Rubber Price', 'Soft sawnwood Price', 'Wood pulp Price']

plt.figure(figsize=(10,10))
materialslist = ['Coarse wool Price','Copra Price', 'Cotton Price', 'Fine wool Price', 'Hard log Price', 'Hard sawnwood Price', 'Hide Price', 'Plywood Price', 'Rubber Price', 'Soft sawnwood Price', 'Wood pulp Price']

for i in range(len(materialslist)):
    plt.subplot(4,3,i+1)
    plt.subplots_adjust( hspace=1, wspace=0.5)
    plt.title(materialslist[i])
    plt.plot(df[materialslist[i]])
    plt.xticks(rotation=90)
plt.suptitle("Raw-Materials price comparision")
```

Podemos observar que o algod√£o e a borracha s√£o de pre√ßos mais baixos, vamos comparar pre√ßos para entender melhor qual √© o mais baixo.

```python
plt.plot(df[['Cotton Price', 'Rubber Price']])
plt.title("Raw-Materials price comparision")
plt.xlabel("Years")
plt.ylabel("Prices")
plt.legend(['Cotton Price', 'Rubber Price'], loc='upper center',bbox_to_anchor=(1.2,1))
```

<strong> algodao ‚òÅÔ∏èüåø</strong> √© mat√©ria-prima de menor pre√ßo nos √∫ltimos anos. <br>
Tanto o algod√£o quanto a borracha s√£o mat√©rias de baixo pre√ßo

#### Q3: Qual mat√©ria-prima tem a maior e menor varia√ß√£o de % de pre√ßo

```python
plt.figure(figsize=(12,12))
for i in range(len(changelist)):
    r = random.random()
    b = random.random()
    g = random.random()
    
    color = (r, g, b)
    plt.subplot(4,3,i+1)
    plt.subplots_adjust( hspace=1, wspace=0.5)
    plt.plot(df[changelist[i]], c=color)
    plt.xticks(rotation=90)
    plt.title(changelist[i])
    plt.xlabel("Year")
    plt.ylabel("% Change")
```
